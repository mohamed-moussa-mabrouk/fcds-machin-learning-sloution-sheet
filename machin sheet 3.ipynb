```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
```

```python
data={'y_axis_point':[4,3,4,4,6,8,4,5,7,3,4,2,4,6,7,8,5,8,7,6,7,6,5,4],
      'x_axis_point':[2,3,3,5,5,5,6,6,6,7,7,8,9,10,10,10,11,11,12,13,13,14,15,15]}
df = pd.DataFrame(data,columns=['x_axis_point','y_axis_point'])
```

### K-Mean n_cluster = 2 

```python
kmeans_1 = KMeans(n_clusters=2).fit(df)
centroid2 = kmeans_1.cluster_centers_
print(centroid2)
```

```python
plt.figure(figsize = [12, 12])
plt.scatter(df['x_axis_point'], df['y_axis_point'], c= kmeans_1.labels_.astype(float), s=25, alpha=0.5)
plt.scatter(centroid2[:, 0], centroid2[:, 1], c='green', s=100)
plt.show()
```

### K-Means n_cluster = 3:-

```python
kmeans_2 = KMeans(n_clusters=3).fit(df)
centroid3 =kmeans_2.cluster_centers_
print(centroid3)
```

```python
plt.figure(figsize = [12, 12])
plt.scatter(df['x_axis_point'], df['y_axis_point'], c= kmeans_2.labels_.astype(float), s=25, alpha=0.5)
plt.scatter(centroid3[:, 0], centroid3[:, 1], c='green', s=100)
plt.show()
```

### K-Mean n-cluster = 4

```python
kmeans_3 = KMeans(n_clusters=4).fit(df)
centroid4 = kmeans_3.cluster_centers_
print(centroid4)
```

```python
plt.figure(figsize = [12, 12])
plt.scatter(df['x_axis_point'], df['y_axis_point'], c= kmeans_3.labels_.astype(float), s=25, alpha=0.5)
plt.scatter(centroid4[:, 0], centroid4[:, 1], c='red', s=100)
plt.show()
```

### K-Ways n_cluster = 2

```python
spectcluster_1 = SpectralClustering(n_clusters=2).fit(df)
labels = spectcluster_1.labels_
plt.figure(figsize = [12, 12])
plt.scatter(df['x_axis_point'], df['y_axis_point'], c=labels)
plt.show()  
```
